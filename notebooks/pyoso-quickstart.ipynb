```python
query = """
SELECT 
    omp.project_name AS projectName,
    tm.sample_date AS date,
    MAX(CASE WHEN tm.metric_name = 'transaction_count' THEN tm.amount END) AS transaction_count,
    MAX(CASE WHEN tm.metric_name = 'gas_fees_sum' THEN tm.amount END) AS gas_fees_sum,
    MAX(CASE WHEN tm.metric_name = 'address_count' THEN tm.amount END) AS address_count
    
FROM timeseries_metrics_by_project_v0 AS tm
JOIN metrics_v0 AS m ON tm.metric_id = m.metric_id
JOIN projects_v1 AS omp ON tm.project_id = omp.project_id
WHERE 
    tm.metric_name IN ('transaction_count', 'gas_fees_sum', 'address_count')
    AND tm.sample_date >= CURRENT_DATE - INTERVAL '3 months'
    AND omp.project_name IN (
        SELECT project_name 
        FROM projects_v1 
        ORDER BY RANDOM() 
        LIMIT 100
    )
GROUP BY omp.project_name, tm.sample_date
ORDER BY omp.project_name, tm.sample_date DESC;

"""

results = client.query(query_1)
results
```
