```python
def get_project_timeseries_metrics(client, project_limit=50):
    # Step 1: Get project list
    project_query = f"""
        SELECT 
            project_name
        FROM projects_v1
        LIMIT {project_limit};
    """
    projects = client.query(project_query)

    all_metrics = []

    # Step 2: Loop through projects and collect metrics
    for project in projects:
        project_name = project['project_name']
        timeseries_query = f"""
            SELECT 
                m.metric_name as metric_name,
                tm.sample_date as date,
                tm.amount as amount,
                '{project_name}' as project_name
            FROM timeseries_metrics_by_project_v0 AS tm
            JOIN metrics_v0 AS m ON tm.metric_id = m.metric_id
            JOIN projects_v1 AS p ON tm.project_id = p.project_id
            WHERE p.project_name = '{project_name}'
            ORDER BY tm.sample_date DESC;
        """
        metrics = client.query(timeseries_query)
        all_metrics.extend(metrics)

    return all_metrics
```

```python
results = get_project_timeseries_metrics(client, project_limit=50)

# Optional: Convert to a pandas DataFrame
import pandas as pd
df = pd.DataFrame(results)
print(df.head())
```
